```python
import pandas as pd
```

```python
diabetes_ds= pd.read_csv('diabetic_data.csv')
#print(diabetes_ds)


#Metl, hiw i think is uselful
columns_to_melt = ['encounter_id', 'patient_nbr', 'race', 'gender', 'age','admission_type_id', 'weight','diabetesMed','insulin']
melted_ds = diabetes_ds.melt(id_vars=columns_to_melt, value_vars=['max_glu_serum', 'A1Cresult'],
                        var_name='Lab results', value_name='Results')

melted_ds
```

```python
#with meds
columns_to_melt = ['encounter_id', 'patient_nbr', 'race', 'gender', 'age','admission_type_id', 'weight','diabetesMed','insulin']
melted_meds = diabetes_ds.melt(id_vars=columns_to_melt, value_vars=['metformin', 'repaglinide','nateglinide','chlorpropamide','glimepiride','acetohexamide','glipizide', 'glyburide','tolbutamide','pioglitazone','rosiglitazone','acarbose','miglitol','troglitazone','tolazamide','examide','citoglipton'],
    var_name='Measurement Type',
    value_name='Value')

melted_meds
```

```python
#to see the "Impact of HbA1c Measurement on Hospital Readmission Rates"

id_columns = ['encounter_id', 'patient_nbr', 'admission_type_id', 'diabetesMed', 'readmitted']

melted_paper = diabetes_ds.melt(
    id_vars=id_columns, 
    value_vars=['A1Cresult', 'max_glu_serum', 'insulin'], 
    var_name='Lab Test', 
    value_name='Result'
)

melted_paper
```

```python
#reorganizes the melted_ds, where the Lab results, max_glu_serum and A1Cresult, will become new columns. 
pivot_ds = melted_ds.pivot(index=['encounter_id', 'patient_nbr', 'race', 'gender', 'age', 'admission_type_id', 'weight', 'diabetesMed', 'insulin'],
    columns='Lab results',
    values='Results')
pivot_ds
```

```python
# performs an aggregation on the diabetes_ds DataFrame for the columns number_diagnoses, num_medications, and num_lab_procedures. 
#It calculates the median, mean, maximum, and minimum for each of these columns.
diabetes_ds.aggregate({"number_diagnoses":['median','mean','max','min'],
                       "num_medications":['median','mean','max','min'],
                       "num_lab_procedures":['median','mean','max','min']})
```

```python


#Grouping by readmission values, and aggregating the mean values of number_diagnoses, num_medications and num_lab_procedures. 


readmission_ds = diabetes_ds.groupby('readmitted').aggregate({
    'number_diagnoses': 'median',
    'num_medications': 'median',
    'num_lab_procedures': 'median',
    
    
})

readmission_ds
```
